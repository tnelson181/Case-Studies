```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('freedom_index.csv')
df=pd.DataFrame(data)

Worldbank=pd.read_excel('Worldbank.xlsx')
wb=pd.DataFrame(Worldbank)
```

```python
wb.head()
```

```python
wb.fillna(0, inplace = True)
```

```python
df.head()
```

```python
#Creating a 10-year growth average column for each country
```

```python
wb.tail()
```

```python
wb['10yr'] = wb[[2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]].mean(axis=1)
```

```python
wb['10yr']
```

```python
wb.head()
```

```python
df= pd.merge(df, wb[['Country Name', '10yr']], how='left', left_on='Country', right_on='Country Name')
```

```python
df.head()
```

```python
cleaned_df = df.dropna()
```

```python
cleaned_df.head()
```

```python
cleaned_df.shape
```

```python
cleaned_df.drop(['Country Name'], axis=1)
```

```python
cleaned_df.drop(['Unnamed: 0'], axis=1)
```

```python
dftop = cleaned_df.sort_values(by='Overall Score', ascending=False).head(20)
```

```python
dfbot = cleaned_df.sort_values(by='Overall Score', ascending=True).head(20)
```

```python
newdf = pd.concat([dftop, dfbot])
```

```python
dftop
```

```python
dfbot.shape
```

```python
newdf
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=newdf, x='Overall Score', y='10yr', color='gray', alpha=0.5)
sns.scatterplot(data=dftop, x='Overall Score', y='10yr', color='blue', label='Top 20')
sns.scatterplot(data=dfbot, x='Overall Score', y='10yr', color='red', label='Bottom 20')

plt.xlabel('Overall Score')
plt.ylabel('10yr')
plt.title('Scatterplot of Overall Score vs 10yr')

plt.legend()

plt.show()
```

```python
# Here we see China's GDP eclipsing its nearest competitor by a factor of 3. China's overall score is the highest of the bottom
# 20, at 48.5. We will review this scatterplot again with China removed, so we can get a better picture of the rest of the nations.
```

```python
newdf = newdf[newdf['Country Name'] != 'China']
```

```python
dfbot = dfbot[dfbot['Country Name'] != 'China']
```

```python
# Now with the outlier of China removed, we look again to compare the two datasets.
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=newdf, x='Overall Score', y='10yr', color='gray', alpha=0.5)
sns.scatterplot(data=dftop, x='Overall Score', y='10yr', color='blue', label='Top 20')
sns.scatterplot(data=dfbot, x='Overall Score', y='10yr', color='red', label='Bottom 20')

plt.xlabel('Overall Score')
plt.ylabel('10yr')
plt.title('Scatterplot of Overall Score vs 10yr')

plt.legend()

plt.show()
```

```python
# As expected, nations with high freedom indexes tend to outperform those with lower scores on an average of their 10 year GDP.
# The difference, however, is surprisingly small compared to expectations.
```

```python
cutoff_point = newdf['Overall Score'].median()

# Step 2: Create a new column indicating whether each row belongs to the top or bottom half
newdf['Top or Bottom'] = 'Top Half'
newdf.loc[df['Overall Score'] <= cutoff_point, 'Top or Bottom'] = 'Bottom Half'

# Step 3: Create the histogram with Seaborn
plt.figure(figsize=(12, 6))
sns.histplot(data=newdf, x='Country Name', y='10yr', hue='Top or Bottom', palette={'Top Half': 'blue', 'Bottom Half': 'red'}, multiple='stack')

plt.xticks(rotation=90)
plt.xlabel('Country Name')
plt.ylabel('10yr')
```

```python
# Here we see an interesting insight - a surprising number of the freest economies are not the largest western economies (aside from
# Germany), but smaller nations. We see that Luxembourg, Estonia, Lithuania, Cyprus, Latvia, Mauritius produce roughly the same as a majority
# of the least free countries, but slightly larger economies like Norway and Sweden are on bar with the oil-rich UAE. Canada, the Netherlands,
# Australia and Switzerland surpass the UAE, with Germany greatly outperforming their 38 competitors. We will, however, now need to again compare
# these results with China's economy
```

```python
dfbot = pd.concat([dfbot, cleaned_df[cleaned_df['Country'] == 'China']])
```

```python
newdf = pd.concat([newdf, cleaned_df[cleaned_df['Country'] == 'China']])
```

```python
newdf.shape
```

```python
cutoff_point = newdf['Overall Score'].median()

newdf['Top or Bottom'] = 'Top Half'
newdf.loc[df['Overall Score'] <= cutoff_point, 'Top or Bottom'] = 'Bottom Half'
plt.figure(figsize=(12, 6))
sns.histplot(data=newdf, x='Country Name', y='10yr', hue='Top or Bottom', palette={'Top Half': 'blue', 'Bottom Half': 'red'}, multiple='stack')
plt.xticks(rotation=90)
plt.xlabel('Country Name')
plt.ylabel('10yr')
```

```python
# Here we gain another important insight by re-adding China - with China removed, the UAE was moved from the top 20 scorers where it belonged,
# into the bottom 20 group. Meaning, that although China is by far the highest performing economy among all 40, aside from China all 19 other
# nations in the bottom 20 were a flat line.
```

```python
# With these insights in mind, we will do a final visualization of the populations of each of these 40 countries, to see if 
# their distributions look the same, of if some of these nations' populations are much different proportionally to one another
# when compared to their 10 year GDP averages.
```

```python
popdata = pd.read_csv('export.csv')
```

```python
CIA = pd.DataFrame(popdata)
```

```python
CIA.head()
```

```python
CIA.rename(columns={'value' : 'population'}, inplace=True)
```

```python
CIA.head()
```

```python
CIA = CIA[['name', 'population']]
```

```python
newdf=pd.merge(newdf, CIA, how='outer', left_on='Country', right_on='name')
```

```python
dftop=pd.merge(dftop, CIA, how='outer', left_on='Country', right_on='name')
```

```python
dfbot=pd.merge(dfbot, CIA, how='outer', left_on='Country', right_on='name')
```

```python
newdf['population'] = newdf['population'].str.replace(',', '')
```

```python
newdf['population'] = newdf['population'].astype(int)
```

```python
cutoff_point = newdf['Overall Score'].median()

newdf['Top or Bottom'] = 'Top Half'
newdf.loc[newdf['Overall Score'] <= cutoff_point, 'Top or Bottom'] = 'Bottom Half'

plt.figure(figsize=(12, 6))
sns.histplot(data=newdf, x='Country Name', y='population', hue='Top or Bottom', palette={'Top Half': 'blue', 'Bottom Half': 'red'}, multiple='stack')
plt.xticks(rotation=90)
plt.xlabel('Country Name')
plt.ylabel('Population')
plt.title('Comparison of populations')

plt.legend(loc='upper left')
plt.show()
```

```python
# Finally, with populations included we can see that the only nation on either side with a vastly larger population is China,
# so the stronger performance of the economically freer nations in terms of GDP does not come from larger populations.
```
